# Columns Summary:

invoice_no: Invoice number associated with each transaction.            

customer_id: Identifier for each customer.

gender: Gender of the customer (assumed to be binary: male/female).

age: Age of the customer.

category: Product category associated with the transaction.

quantity: Quantity of products purchased in each transaction.

selling_price_per_unit: Selling price per unit of the product.

cost_price_per_unit: Cost price per unit of the product.

payment_method: Method used for payment in the transaction.

region: Geographic region associated with the transaction.

state: State where the transaction took place.

shopping_mall: Shopping mall where the transaction occurred.

invoice_time: Time of the invoice (assuming time of day).

invoice_dates: Date of the invoice.

Profit_margin: Profit margin associated with the transaction.

Age_category: Categorized age groups (youth, adults, older, seniors).

Total_Cost: Total cost of the transaction.

## 2.How do sales vary over time?

Use pandas to group sales data by data and matplotlib to visualize the trend


```python
# Convert the 'invoice_date' column to datetime format
df['invoice_date'] = pd.to_datetime(df['invoice_date'])

# Group df by date and calculate the total sales for each date
df= df.groupby(df['invoice_date'].dt.date)['selling_price_per_unit'].sum()

# Plotting the sales trend over time
plt.figure(figsize=(12, 6))
plt.plot(df.index, df.values, marker='o',color='b', linestyle='-')
plt.title('Sales Trend Over Time')
plt.xlabel('invoice_date')
plt.ylabel('Total Selling price')

plt.xticks(rotation=45)

plt.show()

```

# What is the correlation between different numerical variables (e.g., sales ,cost ,profit)?

Use pandas to calculate correlation coefficients and seaborn to create a heatmap.


```python
# Calculate correlation coefficients
df = df[['selling_price_per_unit', 'cost_price_per_unit']].corr()

# Create a heatmap using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(df, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

# Which are the top-selling products?

Use pandas to aggregate sales data by product and visualize the top products with matplotlib or seaborn



```python

# Aggregate sales data by product and calculate the total sales for each product
df = df.groupby(['invoice_no'])['selling_price_per_unit'].sum().reset_index()

# Sort products by total sales in descending order to get the top-selling products
df = df.sort_values(by='selling_price_per_unit', ascending=False).head(10)

# Plotting the top-selling products
plt.figure(figsize=(12, 6))
plt.bar(df['invoice_no'], df['selling_price_per_unit'], color='orange')
plt.title('Top Selling Products')
plt.xlabel('invoice no')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
